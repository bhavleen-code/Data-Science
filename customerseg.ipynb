```python
df_segmentation=pd.read_csv("Converted_data.csv",index_col=0)
```

```python
plt.figure(figsize=(12,9))
plt.scatter(df_segmentation.iloc[:, 2],df_segmentation.iloc[:, 3])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('VISUALISATION OF RAW DATA')
```

```python
scaler=StandardScaler()
segmentation_std= scaler.fit_transform(df_segmentation)
```

```python
hier_clust=linkage(segmentation_std, method='ward')
```

```python
plt.figure(figsize=(12,9))
plt.title('Hirerarchical Clustering Dendrogram')
plt.xlabel("Observations")
plt.ylabel('Distance')
dendrogram(hier_clust,
           truncate_mode='level',
           p=5,
          show_leaf_counts=False,
          no_labels=True,
          color_threshold=0)
plt.show()
```
